```python
import tensorflow as tf #Esta libreria me permite formar la red neuronal
import numpy as np #Esta libreria me permite crear arrays
```

```python
celcius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
#Aqui se especifica el numero de neuronas que deseo que tenga mi red
#capa = tf.keras.layers.Dense(units=1, input_shape=[1]) #Envio una neurona de entrada y una entrada de salida
#modelo = tf.keras.Sequential([capa])

#Entre mas neuronas tenga es mucho mas rapido el aprendimiento
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo =tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
modelo.compile(
    optimizer =tf.keras.optimizers.Adam(0.1), #Entre mas bajo sea este numero de Adam mejor aprendijaze va a tener el sistema
    loss = 'mean_squared_error' #Cantidad de perdida o de errores se tuvo durante el entrenamiento
)
```

```python
print("Comenzando el entrenamiento")
historial = modelo.fit(celcius, fahrenheit, epochs = 1000) #Aqui hago el entrenamiento dandole las variables y la cantidad de epocas o veces que queremos que el sistema repita el entrenamiento
print("Modelo entrenado")
```

```python
#Realizamos la grafica para saber cuanto le tomo al sistema entrenarse por completo
import matplotlib.pyplot as plt
plt.xlabel("#Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
#Probamos la eficacia del sistema entrenado
print("Hacemos una prediccion")
resultado = modelo.predict([100]) #Aqui damos el valor que deseamos que el sistema transforme
print("El resultado es: "+str(resultado) + "Fahrenheit")
```
